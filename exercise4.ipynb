### Activity Selection Problem

Given a set of activities, along with the starting and finishing time of each activity, find the maximum number of activities performed by a single person assuming that a person can only work on a single activity at a time.

For example,

Input: Following set of activities (1, 4), (3, 5), (0, 6), (5, 7), (3, 8), (5, 9), (6, 10), (8, 11), (8, 12), (2, 13), (12, 14)  

Output: (1, 4), (5, 7), (8, 11), (12, 14)

## Tasks

1. Write a program using a recursive greedy approach to get the maximum number of activities (`50 points`)

```python
#RECURSIVE

def maximum_num(activities, selected=None, finish_time=0, index=0):
    if selected is None:
        selected = []

    if index >= len(activities):
        return selected

    starting, ending = activities[index]

    if starting >= finish_time:
        return maximum_num(activities, selected + [(starting, ending)], ending, index + 1)
    else:
        return maximum_num(activities, selected, finish_time, index + 1)

def main():
    activities = [(1, 4), (3, 5), (0, 6), (5, 7), (3, 8), (5, 9), (6, 10), (8, 11), (8, 12), (2, 13), (12, 14)]
    activities.sort(key=lambda x: x[1]) #sort acts
    selected_activities = maximum_num(activities)
    print(selected_activities)

if __name__ == "__main__":
    main()
```

2. Write a program using a iterative greedy approach to get the maximun number of activities (`50 points`)

```python
# ITERATIVE


def iterative_maximum_num(activities):
    activities.sort(key=lambda x: x[1]) #sort compute

    activities_list = []

    time_end = 0

    for start, end in activities:
        if start >= time_end:
            activities_list.append((start, end))
            time_end = end #update endtime

    return activities_list

def main():
    activities = [(1, 4), (3, 5), (0, 6), (5, 7), (3, 8), (5, 9), (6, 10), (8, 11), (8, 12), (2, 13), (12, 14)]
    activities_list = iterative_maximum_num(activities)
    print(activities_list)

if __name__ == "__main__":
    main()
```
