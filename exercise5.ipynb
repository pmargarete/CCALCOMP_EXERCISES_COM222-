### Maximum Subarray
Given an array of integers nums, find the subarray with the largest sum and return the sum.

A subarray is a contiguous non-empty sequence of elements within an array.



Example 1:

Input: nums = [2,-3,4,-2,2,1,-1,4]

Output: 8

Explanation: The subarray [4,-2,2,1,-1,4] has the largest sum 8.

## Tasks

1. Put your student number in a list and manually compute the maximum subarray in your student number (`10 points`)

```python
# getting maximum sub array

def maximum_subarray(stud_num):
  max_found = max_total = stud_num[0]

  for num in stud_num[1:]:
    max_found = max(num, max_found + num)
    if max_found > max_total:
      max_total = max_found

  return max_total

stud_num = [1, 0, 7, 0, 0, 8]
max_sum = maximum_subarray(stud_num)
print("Manual Maximum Sum:", max_sum)
```

2. Write a program using a brute-force approach to get the maximum-subarray of your student number (`40 points`)

```python
# brute max subarray

def brute_subarray_max(my_num):
  n = len(my_num)
  max_sum = 0
  start = end = 0

  for i in range(n):
    current_sum = 0
    for j in range(i, n):
      current_sum += my_num[j]
      if current_sum > max_sum:
        max_sum = current_sum
        start = i
        end = j

  return max_sum

my_num = [1, 0, 7, 0, 0, 8]
max_sum = brute_subarray_max(my_num)
print("Brute Maximum Subarray Sum:", max_sum)
```

3. Write a program using a dynamic programming to get the maximum-subarray of your student number (`50 points`)

```python
# dynamic

def dynamic_max_subarray(my_id):
  max_not_final = my_id[0]
  higher_max = my_id[0]

  for i in range(1, len(my_id)):
    max_not_final = max(my_id[i], max_not_final + my_id[i])
    higher_max = max(higher_max, max_not_final)

  return higher_max

my_id = [1, 0, 7, 0, 0, 8]
max_sum = dynamic_max_subarray(my_id)
print("Dynamic Maximum Subarray Sum:", max_sum)
```
